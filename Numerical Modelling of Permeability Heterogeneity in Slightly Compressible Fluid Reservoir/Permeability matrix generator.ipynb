```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy
import pandas as pd
import statistics
import time
from tqdm import tqdm
import seaborn as sns
%matplotlib qt
```

```python
def af(x,c,p):  # Activation Function
    return 1 if x<c else p
def iaf(x,c,p):  # Inverse Activation Function
    return p if x<c else 1
```

```python
N=41  # Number of blocks in each direction
h1=int(N/4)  # Centers of the oblong ellipses
k1=int(.17*N)
h2=int(.3*N)
k2=int(.5*N)
h3=int(.6*N)
k3=int(.35*N)
perm=np.zeros((N,N))
```

```python
for i in range(N):
    for j in range(N):
        perm[i,j]=80*(np.exp(-(iaf(i,h1,.4)*((i-h1)**2)+(iaf(j,k1,.3))*((j-k1)**2))/25))+130*(np.exp(-(iaf(i,h2,.5)*((i-h2)**2)+(af(j,k2,.35))*((j-k2)**2))/27))+110*(np.exp(-(af(i,h3,.2)*((i-h3)**2)+(iaf(j,k3,.5))*((j-k3)**2))/80))+15*np.random.rand(N,N)[i,j]+30-30*(np.exp(-(af(j,int((N-i)**(.75)+.6*N),.5)*((j-int((N-i)**(.75)+.6*N))**2))/2.5))-27*(np.exp(-(((j-int((N-i)**(.83)+.73*N))**2))/1.5))-27*(np.exp(-(((j-int((N-i)**(.87)+.86*N))**2))/1.5))
ax = sns.heatmap( perm , linewidth = 0.001 , cmap = 'coolwarm' )
  
plt.title( "2D Permeability Heterogeneity" )
plt.show()
```
